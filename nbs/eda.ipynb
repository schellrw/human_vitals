```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/human_vital_signs_dataset_2024.csv')
df.head()
```

```python
df.info()
```

```python
# Check missing values
print("Missing Values:\n", df.isnull().sum())

# Basic statistics
print("\nBasic Statistics:\n", df.describe())
```

```python
risk_counts = df['Risk Category'].value_counts()
print(risk_counts)

risk_counts.plot(kind='bar')
plt.show()

risk_counts.plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# Set up a figure with multiple subplots
plt.figure(figsize=(15, 5))

# Heart Rate Distribution
plt.subplot(131)
sns.histplot(data=df, x='Heart Rate', kde=True)
plt.title('Heart Rate Distribution')

# Respiratory Rate Distribution
plt.subplot(132)
sns.histplot(data=df, x='Respiratory Rate', kde=True)
plt.title('Respiratory Rate Distribution')

# Body Temperature Distribution
plt.subplot(133)
sns.histplot(data=df, x='Body Temperature', kde=True)
plt.title('Body Temperature Distribution')

# # Risk Level Count
# plt.subplot(134)
# sns.countplot(data=df, x='Risk Category')
# plt.title('Risk Level Distribution')
# plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot with risk levels colored
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Heart Rate', y='Respiratory Rate', hue='Risk Category')
plt.title('Heart Rate vs Respiratory Rate by Risk Risk Category')
plt.show()
```
